For Century, clean up all the email addresses. they should all end in @century21.ca
For Royal, standardize the numbers to +11231231234 format (remove dots, add +1)

For both
- take the name column, and split it into first_name, middle_names, last_names columns
- Capitalize the names normally (Bruce Wang)
- create another column called domain which would have the domain name from the email in it

Then finally merge the 2 spreadsheets into one

```python
import os
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
century_orig = pd.read_csv('./century.csv')
royal_orig = pd.read_csv('./royal.csv')
```

```python
century = century_orig
```

```python
royal = royal_orig
```

1. Century Email

```python
fail_list = []
for i in range(len(century)):
    try:
        century.iloc[i] [['email']] = century.iloc[i] [['email']].email.split('@')[0]+'@century21.ca'
    except:
        fail_list.append(i)
print(fail_list)
```

```python
century = century.drop(fail_list)
```

```python
century = century.reset_index(drop=True)
```

2. Royal Phone

```python
royal.iloc[10][['Mobile']].isnull() == True
```

```python
exception_list = []
for i in range(len(royal)):
    if (royal.iloc[i][['Mobile']].isnull().bool() == False):
        phone = royal.iloc[i][['Mobile']].Mobile.replace(' ','').replace('-','').replace('(','').replace(')','').replace('.','')
        if(len(phone)!=10):
            exception_list.append(i)
        else:
            royal.iloc[i] [['Mobile']] = "+1"+phone

print(exception_list)
```
